```python
# Read catalogue listing all available data stores
df = pd.read_csv("https://cmip6.storage.googleapis.com/pangeo-cmip5.csv")
```

```python
# Google Cloud Storage 
# https://gcsfs.readthedocs.io/en/latest/
import gcsfs
fs = gcsfs.GCSFileSystem(token = "anon", access = "read_only")
```

```python
df.columns

# Filtering for historical data
df[(df["source_id"] == "ACCESS1-3") & (df["member_id"] == "r1i1p1f1") & (df["experiment_id"] == "historical")]

# Filtering for RCP85 data (projections)
df[(df["source_id"] == "ACCESS1-3") & (df["member_id"] == "r1i1p1f1") & (df["experiment_id"] == "rcp85")]
```
